# Data analysis questions about the data

```python
#what is the maximum price of diamonds
#print max price
df.price.max()
print("The maximun price of diamonds is {}".format(df.price.max()))
```

```python
#what is the minimum price of diamonds
#print min price
df.price.min()
print("The minimum price of diamonds is {}".format(df.price.min()))
```

```python
#what is the average price of diamonds
#print average price
df.price.mean()
print("The average price of diamonds is {}".format(df.price.mean()))
```

```python
#what is the count of the ideal cut diamonds
# The 'cut' column contains the cut grades of the diamonds.
df[df.cut=='Ideal'].shape[0]
print("The number of diamonds with cut Ideal is {}".format(df[df.cut=='Ideal'].shape[0]))
```

```python
#what is the unique colors of diamonds
#the unique colors of diamonds
df.color.unique()
print("The unique colors of diamonds are {}".format(df.color.unique()))
```

```python

#what is The median of carat in Premium cut
df[df['cut'] == 'Premium']['carat'].median()
print("The median of carat in Premium cut is {}".format(df[df['cut'] == 'Premium']['carat'].median()))
```

```python

#what is carat mean for each cut
df.groupby('cut').carat.mean()  
print("carat mean for each cut is {}".format(df.groupby('cut').carat.mean()))
```

```python
#what is The average table size for each clarity
df.groupby('clarity').table.mean()
print("The average table size for each clarity is: {} ".format(df.groupby('clarity').table.mean()))
```

```python
#what is the number of diamonds with clarest clarity
df[df.clarity=='IF'].count()
print("The number of diamonds with IF clarity is {}".format(df[df.clarity=='IF'].count()))
```

```python
#what is the number of diamonds with dirtiest clarity
df[df.clarity=='I1'].shape[0]
print("The number of diamonds with I1 clarity is {}".format(df[df.clarity=='I1'].shape[0]))
```

```python
#what is the most common clarity of diamonds
df['clarity'].value_counts().idxmax()
print("The most common clarity of diamonds is {}".format(df['clarity'].value_counts().idxmax()))
```

```python
#what is the number of diamonds with each color
df.groupby('color').count()   
print("The number of diamonds with each color is {}".format(df.groupby('color').count()))
```

```python


# Filter the dataset for the cleanest diamonds (clarity 'IF')
cleanest_diamonds = df[df['clarity'] == 'IF']

# Calculate the average price of the cleanest diamonds
average_price_cleanest = cleanest_diamonds['price'].mean()

print(f"The average price of the cleanest diamonds (clarity 'IF') is: {average_price_cleanest}")
```

```python


# The 'clarity' column contains the clarity grades of the diamonds.
# Assuming 'I1' is the lowest clarity grade (dirtiest diamonds):
dirtiest_diamonds = df[df['clarity'] == 'I1']

# Calculate the average price of the dirtiest diamonds
average_price_dirtiest = dirtiest_diamonds['price'].mean()
print(f"The average price of the dirtiest diamonds (clarity 'I1') is: {average_price_dirtiest}")
```

```python
#what is The 'color' column contains the color grades of the diamonds.
df.color.value_counts().idxmax()
print("The most common color of diamonds is {}".format(df.color.value_counts().idxmax()))
```

```python
#what is mean for each color
df.groupby('color').price.mean()
print("price mean for each color is {}".format(df.groupby('color').price.mean()))

```

# Plots and insights

```python
#  Histogram of diamond prices
plt.figure(figsize=(10, 6))
sb.histplot(data=df, x='price', y='carat',alpha = 0.5, bins=20,hue='color', palette='viridis')
plt.title('Distribution of Diamond Prices')
plt.xlabel('Price ($)')
plt.ylabel('Count')
plt.show()

#The graph provides interesting insights into the distribution of diamond prices, the relationship between price and weight and the effect of the color of the diamond on its price.
# It is important to note that these are general insights, and there may be exceptions.
```

```python
#Relationship between Carat and Price
plt.figure(figsize=(10, 6))
sb.scatterplot(data=df,y='carat' , x='price', hue='color', palette='viridis')
sb.regplot(data=df, y='carat', x='price', scatter=None, color='black')
plt.title('Relationship between Carat and Price')
plt.xlabel('Price ($)')
plt.ylabel(' Carat')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title='Color')
plt.show()

#The graph confirms the known relationship between the weight of the diamond and its price, but also highlights the significant effect of the color of the diamond on its price.
# The dispersion of the data suggests examining additional factors that affect the value of the diamond
```

```python
#cut vs. Price Box Plot 
plt.figure(figsize=(10, 6))
sb.lineplot(data=df, x='cut', y='price', hue='color', palette='viridis')
plt.title('cut vs. Price Box Plot')
plt.xlabel('cut')
plt.ylabel('Price ($)')
plt.xticks(rotation=45)
plt.show()

#The graph shows the significant effect of the depth level on the price of the diamond, alongside price differences between different cutss.
#However, other factors also affect the price, and must be taken into account when evaluating diamonds
```

```python
#Average Price by Clarity and Color
plt.figure(figsize=(8, 6))
sb.barplot(data=df, x='clarity', y='price', errorbar=None, hue='color', palette='viridis')
plt.title('Average Price by Clarity')
plt.xlabel('Clarity')
plt.ylabel('Average Price')
plt.show()

#The graph provides insight into the relationship between the level of cleanliness, the color of the diamond and its average price.
# Diamonds with a higher level of clarity and "perfect" color tend to be the most expensive.
```

```python
#Proportion of Diamonds by Color
color_counts = df['color'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(color_counts, labels=color_counts.index, autopct='%1.1f%%',explode=(0.1,0.1,0,0,0,0,0.1),colors=sb.color_palette('viridis', len(color_counts)) )
plt.title('Proportion of Diamonds by Color')
plt.axis('equal')
plt.show()

#Proportion of Diamonds by Color,color G most popular after him E To the nation J that are less popular
```

```python
# Diamond Cut Distribution
plt.figure(figsize=(10, 6))
sb.countplot(data=df, x='cut', palette='viridis')
plt.title('Diamond Cut Distribution')
plt.xlabel('Cut')
plt.ylabel('Count')
for i,j in zip(range(5),df['cut'].value_counts()):    
    plt.annotate(str(j/1000),xy=(i,j),fontsize = 10, color= '#000000',ha = 'center',rotation = 360)
plt.show()

#The graph provides a general picture of the distribution of polishing types in the diamond data.
# There seems to be a preference for high quality polish, but there are also less common types of polish.
```

```python
# Clarity vs. Price Box Plot
plt.figure(figsize=(10,6))
sb.boxenplot(data=df,x='clarity',y='price',palette='viridis')
plt.title('clarity vs.price box plot')
plt.xlabel('clarity')
plt.ylabel('price($)')
plt.xticks(rotation=45)
plt.show()

#The graph shows a clear picture of the average price according to the level of cleanliness This means that the higher the cleanliness,
# the higher the price of the diamond.
```

```python
# Bar Plot of Average Diamond Price by Color
diamonds = sb.load_dataset('diamonds')
mean_prices_by_color = diamonds.groupby('color')['price'].mean().sort_values(ascending=True)
plt.figure(figsize=(10, 6))
sb.barplot(x=mean_prices_by_color.index, y=mean_prices_by_color.values, palette='viridis')
plt.title('Bar Plot of Average Diamond Price by Color')
plt.xlabel('Color')
plt.ylabel('Average Price')
for i,j in zip(range(len(mean_prices_by_color.values)), mean_prices_by_color.values):
    plt.annotate(str(round(j,1)), xy=(i,j), fontsize=10, color='black', fontweight='bold')
plt.show()

#The graph provides insight into the relationship between the color of the diamond and its average price, color j The most expensive,
# and color E the cheapest.
```
